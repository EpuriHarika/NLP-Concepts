# WordNwt With NLTK 

WordNet is a lexical database for the English language is part of the NLTK corpus.

WordNet NLTK module to find the meanings of words, synonyms, antonyms, and more. 



```python
#First, you're going to need to import wordnet:
from nltk.corpus import wordnet
```

```python
#Then, we're going to use the term "program" to find synsets like so:
syns = wordnet.synsets("program")
print(syns[0].name())
```

```python
#For Just the word:
print(syns[0].lemmas()[0].name())
```

```python
#Definition of that synset:
print(syns[0].definition())
```

```python
#Examples of the word in use:
print(syns[0].examples())
```

Next, how might we discern synonyms and antonyms to a word? The lemmas will be synonyms, and then you can use .antonyms to find the antonyms to the lemmas. As such, we can populate some lists like:

```python
synonyms = []
antonyms = []

for syn in wordnet.synsets("happy"):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())
print("Printing synonyms: ")
print(set(synonyms))

print("Printing antonyms")
print(set(antonyms))
```

Next, we can also easily use WordNet to compare the similarity of two words and their tenses, by incorporating the Wu and Palmer method for semantic related-ness.

```python
#Let's compare the noun of "laptop" and "computer":

w1 = wordnet.synset('laptop.n.01')
w2 = wordnet.synset('computer.n.01')

#wup_similarity is for comparing synonyms
print(w1.wup_similarity(w2))
```

```python
#Let's compare the noun of "laptop" and "phone":

w1 = wordnet.synset('laptop.n.01')
w2 = wordnet.synset('phone.n.01')

#wup_similarity is for comparing synonyms
print(w1.wup_similarity(w2))
```
